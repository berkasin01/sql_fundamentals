```python
import sqlite3
import pandas as pd
```

```python
# Load CSV into SQLite
conn = sqlite3.connect('sleep.db')
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df.to_sql('sleep', conn, index=False, if_exists='replace')
```

```python
pd.read_sql('SELECT * FROM sleep LIMIT 5', conn)
```

```python
pd.read_sql('SELECT Occupation, AVG("Quality of Sleep") AS avg_sleep_quality FROM sleep GROUP BY Occupation', conn)
```

```python
pd.read_sql('SELECT "Stress Level", Occupation FROM sleep GROUP BY Occupation', conn)
```

```python
pd.read_sql('SELECT AVG("Sleep Duration"),* FROM sleep GROUP BY Occupation', conn)
```

```python
pd.read_sql('SELECT "Stress Level", "Quality of Sleep", Occupation FROM sleep GROUP BY Occupation', conn)
```

```python
pd.read_sql('SELECT "Stress Level", "Quality of Sleep", Occupation FROM sleep WHERE "Stress Level" >= 7 GROUP BY Occupation', conn)
```

```python
pd.read_sql('SELECT "BMI Category", "Heart Rate" FROM sleep GROUP BY "BMI Category"', conn)
```

```python
pd.read_sql('SELECT Occupation, COUNT(Occupation) AS count_occu FROM sleep GROUP BY Occupation ORDER BY count_occu DESC', conn)
```

```python
pd.read_sql('SELECT "Physical Activity Level", "Quality Of Sleep" Occupation FROM sleep GROUP BY Occupation ORDER BY "Physical Activity Level" DESC ', conn)
```
